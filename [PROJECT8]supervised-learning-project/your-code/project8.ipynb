I just want the column with ladder classification of each country, and as it is not a value, only an index, I will transform it to category.

## Let's begin with IDH as target

Spliting between train and test data:

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_cont_train, y_cont_test = train_test_split(X, y_cont, test_size=0.2, random_state=30)
```

#### Linear Regression

```python
from sklearn import linear_model
linear_countries = linear_model.LinearRegression()
```

```python
linear_countries.fit(X_train, y_cont_train)
```

```python
y_cont_pred = linear_countries.predict(X_test)
```

```python
## Dataframe for comparing predicted y with real y:
a = pd.DataFrame(y_cont_pred).reset_index(drop=True)
b = y_cont_test.reset_index(drop=True)

pd.concat([a,b], axis=1)
```

```python
# Evaluation r2 score
from sklearn.metrics import r2_score
r2_score(y_cont_test, y_cont_pred)
```

#### Can we make is simpler?

```python
from sklearn.feature_selection import RFE
```

```python
selector = RFE(linear_countries, n_features_to_select=4)
```

```python
selector = selector.fit(X_train, y_cont_train)
selector.ranking_
```

```python
rkn = selector.ranking_
col = X.columns

sel_cols = pd.DataFrame(data = [rkn, col]).transpose().sort_values(by=0).head(4)
selected_cols = sel_cols[1].to_list()
selected_cols
```

```python
X_test_reduced = X_test[selected_cols]
X_train_reduced = X_train[selected_cols]
```

```python
linear_countries_reduced = linear_model.LinearRegression()
linear_countries.fit(X_train_reduced, y_cont_train)
y_cont_pred_reduced = linear_countries.predict(X_test_reduced)
r2_score(y_cont_test, y_cont_pred_reduced)
```

As r2_score is close to 1, our prediction can be considered good.

I've tested the number of columns from 20 till 3. 4 was the lowest number with good result.

## And when happiness level is the target?

Preparing the datasets:

```python
X_train, X_test, y_cat_train, y_cat_test = train_test_split(X, y_cat, test_size=0.2, random_state=30)
```

I will also create a reduced dataset to proceed the evaluation only with that 4 more important columns:

```python
#Selecting the 10 most relevant variables for each model:
def selector(model,n):
    selector2 = RFE(model, n_features_to_select=n)
    selector2 = selector2.fit(X, y_cat)

    rkn = selector2.ranking_

    sel_cols2 = pd.DataFrame(data = [rkn, col]).transpose().sort_values(by=0).head(n)
    selected_cols2 = sel_cols[1].to_list()
    selected_cols2

    X_red = X[selected_cols2]
    X_train_red, X_test_red, y_cat_train_red, y_cat_test_red = train_test_split(X_red, y_cat, test_size=0.2, random_state=30)
    
    return(X_train_red, X_test_red, y_cat_train_red, y_cat_test_red)
```

Let's proceed using 4 different methods we learned on class.

### 1. Decision tree

```python
from sklearn.tree import DecisionTreeClassifier
arv= DecisionTreeClassifier()

# Train the model using the training sets
arv.fit(X_train,y_cat_train)

#Predict Output
arv_pred = arv.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_cat_test, arv_pred))
```

#### Reduced dataset:

```python
arv_red= DecisionTreeClassifier()

X_train_red, X_test_red, y_cat_train_red, y_cat_test_red = selector(arv_red,10)

# Train the model using the training sets
arv_red.fit(X_train_red,y_cat_train)

#Predict Output
arv_pred_red = arv_red.predict(X_test_red)
```

```python
print(classification_report(y_cat_test, arv_pred_red))
```

### 2. Logistic regression

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()

# Train the model using the training sets
logreg.fit(X_train,y_cat_train)

#Predict Output
logreg_pred = logreg.predict(X_test)
```

```python
print(classification_report(y_cat_test, logreg_pred))
```

#### Reduced dataset:

```python
logreg_red = LogisticRegression()

X_train_red, X_test_red, y_cat_train_red, y_cat_test_red = selector(logreg, 10)

# Train the model using the training sets
logreg_red.fit(X_train_red,y_cat_train)

#Predict Output
logreg_pred_red = logreg_red.predict(X_test_red)
```

```python
print(classification_report(y_cat_test, logreg_pred_red))
```

### 3. SVM

```python
from sklearn import svm
clf = svm.SVC(gamma = 'scale') #Got s better score than using 'auto'

# Train the model using the training sets
clf.fit(X_train,y_cat_train)

#Predict Output
clf_pred = clf.predict(X_test)
```

```python
print(classification_report(y_cat_test, clf_pred))
```

#### Reduced dataset:

```python
try:
    selector(clf,10)
except:
    print("Not possible to run RFE")
```

### 4. KNN

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)

# Train the model using the training sets
knn.fit(X_train, y_cat_train)

#Predict Output
knn_pred = knn.predict(X_test)
```

```python
print(classification_report(y_cat_test,knn_pred))
```

#### Reduced dataset:

```python
try:
    selector(knn)
except:
    print("Not possible to run RFE")
```
